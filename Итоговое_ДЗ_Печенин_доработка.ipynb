```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pymysql 
```

```python
df_csv = pd.read_csv('russia.index.cpi (statbureau.org).csv')
df_csv = df_csv[df_csv['Год'] > 2017].reset_index()
del df_csv ['index']
df_csv['inflation'] = (df_csv.iloc[:, 1:].mean(axis=1).round(2)) - 100
df_csv
df_xlsx = pd.read_excel('RC_F30_11_2018_T30_11_2023.xlsx')
df_xlsx['data'] = pd.to_datetime(df_xlsx['data'])
df_xlsx['year_month'] = df_xlsx['data'].dt.to_period('M')
df_xlsx = df_xlsx.groupby('year_month').agg({'curs': 'mean'}).reset_index()
df_xlsx.head(10)
```

```python
#Строим графики динамики курса рубля 
```

```python
df_xlsx['year_month'] = df_xlsx['year_month'].dt.to_timestamp()
plt.plot(df_xlsx['year_month'], df_xlsx['curs'], color="r")
plt.title('График динамики курса рубля к доллару')
plt.xlabel("год")
plt.ylabel("курс рубля к доллару")
plt.show()
```

```python
#Строю график инфляции 
```

```python
plt.plot(df_csv['Год'], df_csv['inflation'], color="royalblue")
plt.title('График инфляции рубля за последние 5 лет')
plt.xlabel("год")
plt.ylabel("уровень инфляции")
plt.show()
```
